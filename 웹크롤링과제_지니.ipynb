```python
from bs4 import BeautifulSoup
```

```python
from fake_useragent import UserAgent
import requests
```

```python
url = 'https://www.genie.co.kr/chart/top200'
```

```python
ua = UserAgent()
```

```python
# 헤더 선언
headers = {
'User-Agent': ua.chrome,
'referer': 'https://naver.com'
}
```

```python
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
titles = soup.select('a.title.ellipsis')
artists = soup.select('td.info a.artist.ellipsis')
```

```python
artists
```

```python
# 확인 
# artists
```

```python
# for artist in artists:
#     print(artist.text)
```

```python
musics = []
names = []

for title, artist in zip(titles, artists):
    musics.append(title.get_text().strip())
    names.append(artist.text.strip())
```

```python
import pandas as pd

genie = pd.DataFrame({'Service':'Genie','title':musics,
                      'artist':names})

genie.index = [f'{i}위' for i in range(1,51)]

genie.index.name = '순위'

genie
```

```python
genie.to_excel('geniechart.xlsx')
```
