```python
import platform
import matplotlib as mpl
from matplotlib import font_manager, rc

system = platform.system()
if system == "Windows":
    font_path = "C:/Windows/Fonts/malgun.ttf"
elif system == "Darwin":
    font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc"
elif system == "Linux":
    font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
else:
    font_path = None

if font_path:
    font = font_manager.FontProperties(fname=font_path).get_name()
    mpl.rcParams['font.family'] = font
else:
    print("폰트 경로를 확인해주세요.")

mpl.rcParams['axes.unicode_minus'] = False
```

```python
import numpy as np
import pandas as pd

cctv_data = pd.read_csv('cctv_result.csv')
cctv_data
```

```python
import matplotlib.pyplot as plt
```

```python
# 구별 인구수와 cctv 비율
city = cctv_data['구별']
people = cctv_data['인구수']
cctv = cctv_data['CCTV비율']
cc_data = cctv_data[['구별','CCTV비율']]
c_data = len(city)
index = np.arange(c_data)
```

```python
cc_data
```

```python
cc_data.sort_values(by='CCTV비율',ascending=False)
```

```python
plt.barh(index, cc_data, tick_label=city)

plt.title('구별 인구수 대비 CCTV 비율')
plt.show()
```

```python
plt.barh(index, people, tick_label=city)
plt.colors='k'
plt.barh(index, cctv)
plt.colors='r'
plt.title('구별 인구수 대비 CCTV 비율')
plt.show()
```
