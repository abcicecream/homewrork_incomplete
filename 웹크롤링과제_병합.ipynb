```python
import pandas as pd
```

```python
mel = pd.read_excel('melonchart.xlsx')
mel
```

```python
gen = pd.read_excel('geniechart.xlsx')
gen.drop(columns=['순위'],inplace=True)
```

```python
bug = pd.read_excel('bugschart.xlsx')
bug.drop(columns=['순위'],inplace=True)
```

```python
music_concat = pd.concat([mel,bug,gen],join='outer',axis=1)
music_concat
```

```python
music_concat.set_index('순위',inplace=True)
```

```python
music_concat.to_excel('total.xlsx')
```
